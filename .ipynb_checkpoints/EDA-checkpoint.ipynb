```python
## create new column convert 1, 0 to Yes, No
client_data["churned"] = client_data['churn'].map({1:"Yes", 0:"No"})
```

```python
## conversion object into datetime
client_data["date_activ"] = pd.to_datetime(client_data['date_activ'], format="%d-%m-%Y")
client_data["date_end"] = pd.to_datetime(client_data["date_end"], format="%d-%m-%Y")
client_data["date_last_modify_product"] = pd.to_datetime(client_data["date_last_modify_product"], format="%d-%m-%Y")
client_data["date_renewal"] = pd.to_datetime(client_data["date_renewal"], format="%d-%m-%Y")
```

```python
## plot churn get to know how many consumers are churned and not churned 
plt.figure(figsize=(8,4))
total = len(client_data["churn"])
ax = sns.countplot(data=client_data, x="churned")
for p in ax.patches:
    percentage = "{:.1f}%".format(round(100*(p.get_height()/total),2))
    x = p.get_x()+0.35
    y = p.get_height()-1000
    ax.annotate(percentage, (x,y))
plt.xticks(rotation=0.0)
plt.xlabel("Churn", fontdict={"fontsize":14})
plt.ylabel("Churn")
plt.show()
```

About 10% customer as been churned more than 90% customer not churned

```python
## plot total eletric consumption into level
plt.figure(figsize=(15,7))
def eletric_usage(x):
    if x<200:
        return "Low_electric_consumption(kWh)"
    elif 200<x<2000:
        return "Medium_electric_consumption(kWh)"
    elif 2000<x<10000:
        return "High_electric_consumption(kWh)"
    else:
        return "VeryHigh_eletric_consumption(kWh)"

client_data["eletric_usage"] = client_data["eletric_consumption_12months"].apply(lambda x: eletric_usage(x))
total = len(client_data["eletric_consumption_12months"])
ax = sns.barplot(data=client_data, x="eletric_usage", y="eletric_consumption_12months",errorbar=None)
ax.bar_label(ax.containers[0], fontsize=14)
plt.xlabel("Eletricity Consumption Level", fontdict={"fontsize":14})
plt.ylabel("Eletricity Usage Kwh", fontdict={"fontsize":14})
plt.show()
```

while taking eletric consumption of 12 months data making them into level form we look into the plot so there **Very_High_ Eletric_consumption energy** used more like total `263292` Kwh consumed in 12 Months

```python
plt.figure(figsize=(15,7))
def gas_usage(x):
    if x<8000:
        return "Low_gas_usage"
    elif 8000<x<20000:
        return "Medium_gas_usage"
    elif 20000<x<80000:
        return "High_gas_usage"
    else:
        return "VeryHigh_gas_usage"

client_data["gas_usage"] = client_data["consumption_gas_12months"].apply(lambda x: gas_usage(x))
ax = sns.barplot(data=client_data, x="gas_usage", y="consumption_gas_12months", errorbar=None)
ax.bar_label(ax.containers[0], fontsize=14)
plt.xlabel("Gas Consumption Level")
plt.ylabel("Gas Consumption")
plt.show()
```

while taking Gas consumption of 12 months data making them into level form we look into the plot so there **Very_High_ Gas_consumption** used more like total `529714` gas as been consumed in 12 Months

```python
## create a column where churn 0 stands for retention and 1 stands for churn
client_data["retention"] = client_data["churn"]==0
client_data["churn_ed"] = client_data["churn"]==1
```

```python
## estimate the energy off peak by consumption level
plt.figure(figsize=(15,7))
ax = sns.barplot(data=client_data, x="eletric_usage", y="forecast_price_energy_off_peak", errorbar=None)
plt.bar_label(ax.containers[0], fontsize=12)
plt.xlabel("Eletric Consumption Level",fontdict={'fontsize':14})
plt.ylabel("Eletric Price OFF Peak", fontdict={'fontsize':14})
plt.show()
```

 - eletric usage or consumption level of 12months by predict price energy in off peak hours where **Medium Eletric Consumption** 14% energy will affordable due to off peak hours were we use less

```python
## estimate the energy peak by consumption level
plt.figure(figsize=(15,5))
ax = sns.barplot(data=client_data, x="eletric_usage", y="forecast_price_energy_peak",errorbar=None)
plt.bar_label(ax.containers[0])
plt.xlabel("Eletric Consumption Level", fontdict={'fontsize':14})
plt.ylabel("Eletric Price Peak",fontdict={'fontsize':14})

plt.show()
```

- Eletric Usage level by predict the price of energy in peak hours like early morning time where **Very_High_eletric_consumption** will be high 0.6% and more demand in price compare to off peak hours
