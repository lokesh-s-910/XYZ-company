### Feature Engineering

```python
## convert price object into datetime
price_data["price_date"] = pd.to_datetime(price_data["price_date"],format="%d-%m-%Y")
price_data["price_date"].value_counts()
```

```python
## group the company monthly off peak price fixed and variable
monthly_price_id = price_data.groupby(["id", "price_date"]).agg({'price_off_peak_var':'mean','price_off_peak_fix':'mean'}).reset_index()

## extract the month price of jan and dec
jan_price = monthly_price_id.groupby(["id"]).first().reset_index()
dec_price = monthly_price_id.groupby(["id"]).last().reset_index()

## merge the two dataframe jan and dec on id
diff1 = pd.merge(dec_price.rename(columns={"price_off_peak_var":"dec_1_offpeak","price_off_peak_fix":"dec_2_offpeak"}), jan_price.drop(columns="price_date"), on="id")
## diff between jan and dec dataframe off peak var and fixed price
diff1["off_peak_price_dec_jan_energy"] = diff1["dec_1_offpeak"] - diff1["price_off_peak_var"]
diff1["off_peak_price_dec_jan_power"] = diff1["dec_2_offpeak"] - diff1["price_off_peak_fix"]
diff1 = diff1[["id","off_peak_price_dec_jan_energy","off_peak_price_dec_jan_power"]]

## merge with main dataframe
client_data = pd.merge(left=client_data, right=diff1, on="id")
df = client_data[['off_peak_price_dec_jan_energy','off_peak_price_dec_jan_power']]
df.head()
```

```python
## group the company monthly peak price fixed and variable
monthly_price_id = price_data.groupby(["id", "price_date"]).agg({'price_peak_var':'mean','price_peak_fix':'mean'}).reset_index()

## extract the month price of jan and dec
jan_price = monthly_price_id.groupby(["id"]).first().reset_index()
dec_price = monthly_price_id.groupby(["id"]).last().reset_index()

## merge the two dataframe jan and dec on id
diff = pd.merge(dec_price.rename(columns={"price_peak_var":"dec_1","price_peak_fix":"dec_2"}), jan_price.drop(columns="price_date"), on="id")
## diff between jan and dec dataframe peak var and fixed price
diff["peak_price_dec_jan_energy"] = diff["dec_1"] - diff["price_peak_var"]
diff["peak_price_dec_jan_power"] = diff["dec_2"] - diff["price_peak_fix"]
diff = diff[["id","peak_price_dec_jan_energy","peak_price_dec_jan_power"]]

## merge with main dataframe
client_data = pd.merge(left=client_data, right=diff, on="id")
df = client_data[['peak_price_dec_jan_energy','peak_price_dec_jan_power']]
df.head()
```

```python
## taking average on monthwise where energy is peak or off peak
mean_price = price_data.groupby(["id","price_date"]).agg({'price_off_peak_var':'mean',
                                                          'price_peak_var':'mean',
                                                          'price_off_peak_fix':'mean',
                                                          'price_peak_fix':'mean',
                                                         }).reset_index()

## make the diff between peak_price and off peak price variable and fixed
mean_price['offpeak_peak_var_diff'] = mean_price['price_off_peak_var'] - mean_price['price_peak_var']
mean_price['offpeak_peak_fix_diff'] = mean_price['price_off_peak_fix'] - mean_price['price_peak_fix']

## takeing the avg from fixed and varible price
mean_peak_offpeak = mean_price.groupby(['id']).agg({'offpeak_peak_var_diff':"mean",'offpeak_peak_fix_diff':"mean"}
                                            ).reset_index().rename(columns={'offpeak_peak_var_diff':'avg_offpeak_peak_var_diff',
                                                                            'offpeak_peak_fix_diff':'avg_offpeak_peak_fix_diff'}
                                                                   )
mean_peak_offpeak.head()

##merge mean of peak and off peak value to the main dataframe
client_data = pd.merge(left=client_data, right=mean_peak_offpeak, on='id')
```

```python
## find the tenure from consumer active date by end date
client_data['tenure'] = round(abs((client_data['date_activ'] - client_data['date_end'])/pd.to_timedelta(1,"D")/365)).astype("int")
churn_rate = client_data.groupby(['tenure']).agg({'churn':'mean'}).sort_values(by="churn", ascending=False)
churn_rate['churn'] = round(churn_rate['churn']*100).astype('str') + "%"
churn_rate
```

- Most customer planned to get converted like move with other companies who provide service with in **3 Years** customers  planned it to be churned because of the contract they have signed they can't move to the other companies   

```python
## function to convert date into months
def converts_date_to_months(reference_data, data, columns):
    year_to_month = reference_data - data[columns]
    days = (year_to_month/np.timedelta64(1,"D")).astype(int)
    months = round(abs(days/30.417)).astype(int)
    return months
```

```python
## function create date into year
def convert_date_into_year(reference_date, data, columns):
    date = reference_date - data[columns]
    days = (date/np.timedelta64(1,"D")).astype(int)
    year = round(abs(days/365)).astype(int)
    return year
```

```python
from datetime import datetime
reference_date = datetime(2016,1,1)
##function to convert date into years
client_data['contract_date_to_year'] = convert_date_into_year(reference_date, client_data, 'date_activ')
client_data['last_date_modify_prod_in_year'] = convert_date_into_year(reference_date, client_data, 'date_last_modify_product')
## function to convert date into years
client_data['contract_end_date_to_months'] = converts_date_to_months(reference_date, client_data, 'date_end')
client_data['contract_renewal_date_to_months'] = converts_date_to_months(reference_date, client_data, 'date_renewal')

## view data
df = client_data[['contract_date_to_year',
                  'last_date_modify_prod_in_year',
                  'contract_end_date_to_months',
                  'contract_renewal_date_to_months']]
df.head()
```

### Model Building

```python
data = client_data.copy()
data.head()
```

```python
## train and test model and split the model into two parts
from sklearn.model_selection import train_test_split
train_split, test_split = train_test_split(data, train_size=0.7, test_size=0.3, random_state=50)
print("Train Shape: ",train_split.shape)
print("Test Shape: ",test_split.shape)
```

```python
## scale data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
val = ['eletric_consumption_12months',
       'consumption_gas_12months',
       'eletricity_consumption_last_month',
       'forecast_eletric_consump_next_12m',
       'forecast_eletric_consump_nextyear',
       'forecast_discount_energy',
       'forecast_meter_rent_next12month',
       'forecast_price_energy_off_peak',
       'forecast_price_energy_peak',
       'forecast_price_power_off_peak',
       'current_paid_consump',
       'margin_gross_pow_ele',
       'margin_net_pow_ele',
       'nb_prod_act',
       'net_margin',
       'pow_max']

## scale value in train and test model
train_split[val] = scaler.fit_transform(train_split[val])
test_split[val] = scaler.transform(test_split[val])
```

```python
## extract train value into x_train and y_train 
x_train = train_split
y_train = train_split.pop('churn')

x_test = test_split
y_test = test_split.pop('churn')
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=20, max_depth=5, max_features=10, oob_score=True, random_state=65)
random_forest.fit(x_train, y_train)
```

### Plot ROC Curve
- it shows the tradeoff between sensitivity and specifity 
- The closer the curve follows the left-hand border and then the top border of the ROC space, the more accurate the test.
- The closer the curve comes to the 45-degree diagonal of the ROC space, the less accurate the test.

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(estimator=random_forest, X=x_train, y=y_train)
```

- we are not getting perfect auc score which is to less 0.74

### Hyperparameter Coarse Tuning 

```python
%%time
from sklearn.model_selection import GridSearchCV
### tuning random forest using cross validation
random_forest_tuning = RandomForestClassifier(random_state=50, n_jobs=-1)
tuning = {'max_depth':[3,5,7,9,10],
          'max_features':[5,8,12,18,20],
          'min_samples_leaf':[20,35,40,45,80],
          'n_estimators':[15,25,40,55,70]
        }

gridsearch_model = GridSearchCV(estimator=random_forest_tuning,
                             param_grid=tuning,
                             verbose=1,
                             n_jobs=-1,
                             cv=5,
                             return_train_score=True)

## fit the model
gridsearch_model.fit(x_train, y_train)
```

```python
## accuracy of the tuned model
print('Accuracy Coarse Tuned model: ', gridsearch_model.best_score_)
print('-'*100)
## after tuning get the best estimator values for predicting
print("Tuned_Best_Estimator: ", gridsearch_model.best_estimator_)
print("-"*100)
```

```python
## create Dataframe for tuned results
tuned_result = pd.DataFrame(gridsearch_model.cv_results_)
tuned_result.head()
```

```python
## take the important data values from tuned result
imp_tuned_result = ['param_max_depth',
                    'param_max_features',
                    'param_min_samples_leaf',
                    'param_n_estimators',
                    'rank_test_score',
                    'mean_train_score',
                    'mean_test_score']

imp_tuned_values = pd.DataFrame(tuned_result[imp_tuned_result])
imp_tuned_values.sort_values(by='rank_test_score', ascending=False).head()
```

- Model balanced where mean train score and mean test score is equal 
- cross varify with fine tuned model

### Fine Tuning Random Forest Model 

```python
%%time
random_forest_finetuned_model = RandomForestClassifier(max_depth=10, random_state=50, n_jobs=-1)
fine_tuned = {'min_samples_leaf':[10,20,25,30,45],
              'max_features':[2,4,8,9,13],
              'n_estimators':[10,20,30,45,50]}

gridsearch_fine_tuned = GridSearchCV(estimator=random_forest_finetuned_model,
                                     param_grid=fine_tuned,
                                     verbose=1,
                                     n_jobs=-1,
                                     return_train_score=True,
                                     cv=5)


## fit the model
gridsearch_fine_tuned.fit(x_train, y_train)
```

```python
## accuracy of the fine_tuned model
print('Accuracy Fine Tuned model: ', gridsearch_fine_tuned.best_score_)
print('-'*100)
## after tuning get the best estimator values for predicting
print("Fine_Tuned_Best_Estimator: ", gridsearch_fine_tuned.best_estimator_)
print("-"*100)
```
